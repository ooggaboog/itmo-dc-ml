```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# загрузка данных без заголовка
data = pd.read_csv('1-data.csv', header=None)
data
```

```python
# создание экземпляра PCA с кол-вом ГК равным 10 (по кол-ву признаков)
pca = PCA(n_components=10, svd_solver='full')

# применение PCA к данным
pca.fit(data)

# преобразование данных в новые координаты
transformed_data = pca.transform(data)
transformed_data
```

```python
# координаты первого объекта относительно первой и второй ГК
print(round(transformed_data[0][0], 2), round(transformed_data[0][1], 2), sep=', ')
```

```python
# доля объясненной дисперсии при использовании первых двух ГК
round(pca.explained_variance_ratio_[:2].sum(), 2)
```

```python
# минимальное кол-во ГК для доли объясненной дисперсии более 0.85
(pca.explained_variance_ratio_.cumsum() > 0.85).argmax() + 1
```

```python
# количество групп объектов при использовании первых двух ГК
# создание экземпляра KMeans с количеством кластеров от 1 до 10
distortions = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(transformed_data[:,:2])
    distortions.append(kmeans.inertia_)

# визуализация методом локтя для определения оптимального количества кластеров
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.title('Elbow Method')
plt.show()
```
