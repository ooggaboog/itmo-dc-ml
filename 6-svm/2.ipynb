```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('persons_pics_train.csv')
df_reserved = pd.read_csv('persons_pics_reserved.csv')

X_train = df.drop(columns=['label'])
y_train = df['label']

X_test = df_reserved
```

```python
tuned_parameters = [{'kernel': ['linear', 'poly', 'rbf', 'sigmoid'], 'gamma': [1e-3, 1e-4],
                     'C': [1, 10, 100, 1000], 'class_weight': [None, 'balanced']}]


cv = GridSearchCV(SVC(), tuned_parameters, refit=True, verbose=3)

cv.fit(X_train, y_train)
```

```python
cv.best_score_
```

```python
cv.best_params_
```

```python
print(list(cv.predict(X_test)))
```
